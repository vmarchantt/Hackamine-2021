# Modelo de Regresión

```python
from sklearn.model_selection import train_test_split
t = range(len(df_reemplazo["HORAS ACUMULADAS"].dropna()))
x = df_reemplazo["HORAS ACUMULADAS"].dropna()
```

```python
# Separamos datos en conjunto de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(t, x, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg = reg.fit(np.reshape(X_train, (-1,1)), y_train)
```

```python
x_pred = reg.predict(np.reshape(X_test,(-1,1)))
```

```python
plt.figure(figsize=(10,8))
plt.plot(X_test,x_pred, label="Regresión Lineal",color="Red")
plt.scatter(t,x, s=10, label='Data')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r2_score(y_test,x_pred)
```

1.672% de la varianza de los datos queda explicada por el modelo lineal :(

```python
df_minecare = pd.read_csv("Datos/minecare_data_hackamine.csv")
```

```python
df_minecare[df_minecare["oem_id"]!="0"]["id_truck"].unique()
# id truck, 
```

```python
df_minecare[df_minecare["oem_id"]!="0"]
```

```python
df_minecare_2 = pd.DataFrame()
df_minecare_2["id_truck"] = df_minecare[df_minecare["oem_id"]!="0"]["id_truck"][:100000]
df_minecare_2["label"] = df_minecare[df_minecare["oem_id"]!="0"]["float_value"][:100000]
```

```python
df_minecare_2
```

```python
#Pasamos a enteros
label = []
for i in df_minecare_2["label"]:
    label.append(int(i))
```

```python
label = [0 if i<3 else i for i in label]
label = [1 if i>=3 else i for i in label]
```

```python
df_minecare_2["label"] = label
```

```python
# Categorización de conjunto X
df_minecare_2
```

```python
from sklearn.preprocessing import OneHotEncoder

le = OneHotEncoder(sparse=False)
X = le.fit_transform(np.reshape(df_minecare_2["id_truck"].values,(-1,1)))
```

```python
y = df_minecare_2["label"]
```

```python
X.shape
```

```python
# Clasificación mediante KNN

from sklearn.neighbors import KNeighborsClassifier

# Separamos datos en conjunto de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)

```

```python
y_pred = neigh.predict(X_test)
```

```python
# Medidas para evaluar modelo

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
